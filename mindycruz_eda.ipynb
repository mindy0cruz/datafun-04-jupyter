## Import Dependencies:
Imports needed for project

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Data Acquisition:
Load dataset needed for project

```python
# Load the Iris dataset into DataFrame
df = sns.load_dataset('iris')

# Inspect first rows of the DataFrame
print(df.head())
```

## Data Inspection:
Display the first 10 rows of the DataFrame, check the shape, and display the data types of each column using

```python
print(df.head(10))
print(df.shape)
print(df.dtypes)
```

## Descriptive Statistics:
Display summary statistics for each column

```python
print(df.describe())
```

## Data Distribution:
Numerical Columns

```python
# Inspect histogram by numerical column
df['sepal_length'].hist()
df['petal_length'].hist()

# Inspect histograms for all numerical columns
df.hist()

# Show all plots
plt.show()
```

## Data Analysis:
The above data gives us a visual representation of iris sepal and petal lengths and widths. We can see with the overlayed graph that there may be some correlation petal and sepal lengths.


## Data Distribution:
Categorical Columns

```python
# Inspect value counts by categorical column
df['species'].value_counts()

# Inspect value counts for all categorical columns
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()

# Show all plots
plt.show()
```

## Species Anaylsis:
From the abouve graph we can see there was an even distribution of species.

## Data Transformation:


```python
# Renaming a column
df.rename(columns={'sepal_length': 'Sepal Length'}, inplace=True)
df.rename(columns={'sepal_width': 'Sepal Width'}, inplace=True)
df.rename(columns={ 'petal_length': 'Petal Length'}, inplace=True)
df.rename(columns={ 'petal_width': 'Petal Width'}, inplace=True)

# Adding a new column
df['Sepal Area'] = df['Sepal Length'] * df['Sepal Width']
df['Petal Area'] = df['Petal Length'] * df['Petal Width']

```

## Visualizations:

```python
sns.pairplot(df, hue='species')
plt.show()
```

## Conclusion:
We can see there there is correlations between sepal length and petal length.  There are consitencies to be seen across each species.
